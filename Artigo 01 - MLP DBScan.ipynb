```r
mykmeans <- function(data, n_clusters){
    km <- kmeans(data[,1:2], centers=n_clusters)
    return(list("cluster" = km$cluster, "means" = km$centers))
}
```

#### DBScan

```r
mydbscan <- function(data, eps, minPts){
    n_dims <- dim(data)[2]-1
    
    results <- dbscan(data[,1:2], eps = eps, minPts = minPts, weights = NULL, borderPoints = TRUE)
    n_clusters <- max(results$cluster)
    
    cluster_means <- matrix(0, nrow=n_clusters, ncol=n_dims)
    for (k in seq(1,n_clusters)){
        cluster_means[k,] <- colMeans(data[results$cluster==k,1:2])
    }
    return(list("cluster" = results$cluster, "means" = cluster_means))
}
```

```r
dbscan_results <- mydbscan(data = data1, eps = 0.5, minPts = 4)
dbscan_results$means
```

#### Identifying Candidate Points

```r
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
```

```r
myalgorithm <- function(data, eps, tol_perc = 0.001){
    n <- dim(data)[1]
    n_dims <- dim(data)[2] -1 
    n_classes <- max(data[,n_dims+1]) + 1
    candidate <- matrix(0,nrow=n,ncol=1)
    new_class <- matrix(0,nrow=n,ncol=1)
    
    d <- as.matrix(dist(data[,1:2]))
    
    for (i in seq(1,n)){
        nb <- which(d[i,] < eps)
        nb_classes <- data[nb, n_dims+1]
        
        diff_nb <- nb[nb_classes != data[i, n_dims+1]]
        eq_nb <- nb[nb_classes == data[i, n_dims+1]]

        diff_len <- length(as.vector(diff_nb))
        eq_len <- length(as.vector(eq_nb)) - 1 # removing itself
        total_len <- length(as.vector(nb)) - 1 # removing itself
        
        if(total_len == 0){
            candidate[i,1] <- 0
        } else {
            candidate[i,1] <- diff_len / total_len
        }
        #candidate[i,1] <-  1 * length(as.vector(diff_nb)) > length(as.vector(eq_nb))
        #candidate[diff_nb,1] <- 1
        
        # If the observation if close to only observations of a different class, switches its class to the mode class
        if(candidate[i,1] == 1.0){
            new_class[i,1] <- getmode(nb_classes[nb_classes!=data[i, n_dims+1]]) # Switches class if isolate from equals
        } else if (candidate[i,1] >= tol_perc){
            new_class[i,1] <- n_classes + 1 # Assign new class to dubious observations
        } else {
            new_class[i,1] <- data[i,n_dims + 1] # Keeps class if not a candidate or not in tol range
        }
    }
    
    new_data <- data 
    new_data[,n_dims+1] <- new_class
    new_data <- new_data[new_class != n_classes + 1, ]
    
    return(list("candidate" = candidate, "new_class" = new_class, "new_data" = new_data))
}
```

```r
plot_results <- function(data, candidate, new_class, xlim, ylim){
    n_dims <- dim(data)[2]-1
    par(mfrow=c(2,2))
    plot(data$x1, data$x2, col=data$class + 1, xlim = xlim, ylim=ylim, main="Original Data")
    par(new=F)
    plot(data$x1, data$x2, col=data$class + 1, xlim = xlim, ylim=ylim, xlab="", ylab="", main="Observations that Switch Class")
    par(new=T)
    plot(data[candidate==1,1:2], col="blue", pch=4, lwd=2, xlim = xlim, ylim=ylim)
    par(new=F)
    #new_data <- data
    #new_data[candidate==1,n_dims + 1] <- 1 * !as.logical(new_data[candidate==1, n_dims + 1])
    #plot(new_data$x1, new_data$x2, col=new_data$class + 1, xlim = xlim, ylim=ylim, xlab="", ylab="")
    plot(data$x1, data$x2, col=new_class + 1, xlim = xlim, ylim=ylim, xlab="", ylab="", main="High Mixture Region")
    par(new=F)
    last_class <- max(new_class)
    plot(data[new_class != last_class,1], data[new_class != last_class,2], col=new_class[new_class != last_class,1] + 1, xlim = xlim, ylim=ylim, xlab="", ylab="", main="Result")
}
```

# Train Test Split

```r
my_tts <- function(data, train_perc=0.7, shuffle = TRUE){
    
    n <- dim(data)[1]
    n_dims <- dim(data)[2]-1
    
    n_train <- round(n*train_perc)
    n_test <- n - n_train
    if(shuffle){
        train_idx <- sample(n)[1:n_train]
    }
    
    xtrain <- data[train_idx,1:n_dims]
    xtest <- data[-train_idx,1:n_dims]
    ytrain <- data[train_idx,n_dims+1]
    ytest <- data[-train_idx,n_dims+1]
    
    return(list("xtrain" = xtrain, "xtest" = xtest, "ytrain" = ytrain, "ytest" = ytest, "train_idx" = train_idx))
}
```
